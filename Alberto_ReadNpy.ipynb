```python
from google.colab import drive
drive.mount('/content/drive')
 

#!ls
 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import time
import os

from matplotlib import ticker, cm
import matplotlib.colors as colors

cmap = mpl.cm.jet
cmap.set_under('w')
cmap.set_bad('w')

charge2kev = (25.4/38.06) # at 5 GeV with 100 um of silicon, run 46

# nclu = 10
# dt = np.dtype([('Timepix3_0_nclusters','i'),('Timepix3_0_pos_x','f',(nclu,)),('Timepix3_0_pos_y','f',(nclu,)),('Timepix3_0_pos_z','f',(nclu,)),('Timepix3_0_charge','f',(nclu,))])

def func(finp,dist,runnumber):



  #finp = "converted_000061.npy"
  #dist = 405. # mm
  %cd /content/drive/Shareddrives/BL4S/AnalisiBari/Data
  d = np.load(finp)

  print ("Number of events = ", len(d))

  fig0,ax0= plt.subplots(1)
  hist0, bins0, patch0 = ax0.hist(d['Timepix3_0_nclusters'],bins=11,range=[-0.5, 10.5])

  ax0.set_xlabel("Number of Clusters")
  ax0.set_ylabel("Number of Events")
  ax0.set_yscale("log")

  print ("hist0 ", hist0)
  print ("bins0 ", bins0)
  Nele = hist0.sum() - hist0[0]
  print ("Number of Electrons ", Nele)

  TRXray = hist0[1:]/Nele

  Nph = np.arange(0., 10., 1)
  print ("Nph = ", Nph)


  print ("Number of TR X-ray for electrons ", TRXray, (TRXray*Nph).sum())

  fig0,ax0= plt.subplots(1)
  ax0.bar(Nph, TRXray)
  ax0.set_xlabel("Number of TR X-ray") 
  ax0.set_ylabel("Number of TR X-ray/Electron")
  ax0.set_yscale("log")
  textstr = 'Mean %.4f ' % ((TRXray*Nph).sum())
  props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
  ax0.text(0.6, 0.95, textstr, transform=ax0.transAxes, fontsize=10, verticalalignment='top', bbox=props)
                  
  m0 = d["Timepix3_0_nclusters"]>0

  m2 = d["Timepix3_0_nclusters"]==2

  print ("Number of events with two clusters = ", len(d[m2]))
  %cd /content/drive/Shareddrives/BL4S/AnalisiBari/Scripts/AlbertoAnelli
  os.system("mkdir 0000{}".format(runnumber))

  plt.savefig(finp[10:16]+"-0.png")
        
  fig1,ax1= plt.subplots(1)
  cs1 = ax1.hist2d(d['Timepix3_0_pos_x'][m2].T[1]-d['Timepix3_0_pos_x'][m2].T[0],d['Timepix3_0_pos_y'][m2].T[1]-d['Timepix3_0_pos_y'][m2].T[0],bins=200,range=([-4.,4.],[-4.,4.]),cmap=cmap,cmin=1)
  cb1 = fig1.colorbar(cs1[3], ax=ax1)
  cb1.set_label(r"Number of Events/pixel")
  #cb1.set_ticks(ticks, update_ticks=True)
  ax1.set_xlabel(r"X $-$ X$_{part}$ (mm)")
  ax1.set_ylabel(r"Y $-$ Y$_{part}$ (mm)")
  ax1.grid(True)

  Rang = np.sqrt( np.power(d['Timepix3_0_pos_x'][m2].T[1]-d['Timepix3_0_pos_x'][m2].T[0], 2) + np.power(d['Timepix3_0_pos_y'][m2].T[1]-d['Timepix3_0_pos_y'][m2].T[0], 2) )/dist*1.e3 # angular distribution in mrad


  dhist, dbins = np.histogram(Rang, bins=200, range=[0., 10])
  dhist = dhist/Nele
  width = dbins[1] - dbins[0]
  plt.savefig(finp[10:16]+"-1.png")
  fig2,ax2= plt.subplots(1)
  ax2.bar(dbins[:-1], dhist, width=width)
  ax2.set_xlabel("TR X-ray angle (mrad)")
  ax2.set_ylabel("Number of TR X-ray/Electron")
  ax2.set_yscale("log")

  m1 = d["Timepix3_0_nclusters"] > 1

  m11 = d["Timepix3_0_nclusters"] >= 1

  m111 = d["Timepix3_0_charge"]>0

  hEXchar, hEXbins = np.histogram(d['Timepix3_0_charge'][m111].T*charge2kev, bins=50, range=[0, 100])
  hEXchar = hEXchar/Nele
  wx = hEXbins[1] - hEXbins[0]
  hEchar, hEbins = np.histogram(d['Timepix3_0_charge'][m11].T[0]*charge2kev, bins=50, range=[0, 100])
  hEchar = hEchar/Nele
  we = hEbins[1] - hEbins[0]
  hXchar = hEXchar - hEchar
  plt.savefig(finp[10:16]+"-2.png")
  fig3,ax3= plt.subplots(1)
  ax3.bar(hEXbins[:-1], hEXchar, width=we, label="All", alpha=0.5)
  ax3.bar(hEbins[:-1], hEchar, width=wx, label="Electrons", alpha=0.5)
  ax3.bar(hEbins[:-1], hXchar, width=wx, label="TR X-Rays", alpha=0.5)
  ax3.set_xlabel("Energy (keV)")
  ax3.set_ylabel("Fraction of Events")
  ax3.legend()

  
  plt.savefig(finp[10:16]+"-3.png")
  plt.show()

for runnumber in range(109,114):
  finp = "converted_000{}.npy".format(runnumber)
  dist = 40. # mm
  func(finp,dist,runnumber)


```
